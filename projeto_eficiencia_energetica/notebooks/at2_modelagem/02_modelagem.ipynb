# **2. Implementação dos Modelos de Regressão** 

**Agora, vamos gerar os modelos para a implementação de cinco modelos de regressão diferentes:**

```python
# Notebook: at2_modelagem/02_modelagem.ipynb

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import cross_val_score, KFold
import joblib
import time

# Carregar os dados pré-processados
X_train = np.load('../../data/processed/X_train_preprocessed.npy')
X_test = np.load('../../data/processed/X_test_preprocessed.npy')
y_heating_train = np.load('../../data/processed/y_heating_train.npy')
y_heating_test = np.load('../../data/processed/y_heating_test.npy')
y_cooling_train = np.load('../../data/processed/y_cooling_train.npy')
y_cooling_test = np.load('../../data/processed/y_cooling_test.npy')

print("Dados carregados com sucesso.")
print(f"Dimensões do X_train: {X_train.shape}")
print(f"Dimensões do X_test: {X_test.shape}")

# Função para avaliar modelos
def avaliar_modelo(modelo, X_train, X_test, y_train, y_test, nome_modelo, target_name):
    # Medir o tempo de treinamento
    inicio = time.time()
    modelo.fit(X_train, y_train)
    fim = time.time()
    tempo_treino = fim - inicio
    
    # Fazer previsões
    y_pred_train = modelo.predict(X_train)
    y_pred_test = modelo.predict(X_test)
    
    # Calcular métricas
    rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
    mae_train = mean_absolute_error(y_train, y_pred_train)
    mae_test = mean_absolute_error(y_test, y_pred_test)
    r2_train = r2_score(y_train, y_pred_train)
    r2_test = r2_score(y_test, y_pred_test)
    
    # Validação cruzada (5-fold)
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    cv_rmse = np.sqrt(-cross_val_score(modelo, X_train, y_train, 
                                       scoring='neg_mean_squared_error', 
                                       cv=kf)).mean()
    
    # Salvar resultados em um dicionário
    resultados = {
        'Modelo': nome_modelo,
        'Target': target_name,
        'RMSE_Train': rmse_train,
        'RMSE_Test': rmse_test,
        'MAE_Train': mae_train,
        'MAE_Test': mae_test,
        'R2_Train': r2_train,
        'R2_Test': r2_test,
        'CV_RMSE': cv_rmse,
        'Tempo_Treino': tempo_treino
    }
    
    # Salvar o modelo treinado
    joblib.dump(modelo, f'../../models/{nome_modelo}_{target_name}.pkl')
    
    return resultados

# Lista de modelos para testar
modelos_heating = [
    (LinearRegression(), "LinearRegression"),
    (Ridge(alpha=1.0), "Ridge"),
    (Lasso(alpha=0.1), "Lasso"),
    (RandomForestRegressor(n_estimators=100, random_state=42), "RandomForest"),
    (GradientBoostingRegressor(n_estimators=100, random_state=42), "GradientBoosting"),
    (XGBRegressor(n_estimators=100, random_state=42), "XGBoost"),
    (MLPRegressor(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=42), "NeuralNetwork")
]

modelos_cooling = [
    (LinearRegression(), "LinearRegression"),
    (Ridge(alpha=1.0), "Ridge"),
    (Lasso(alpha=0.1), "Lasso"),
    (RandomForestRegressor(n_estimators=100, random_state=42), "RandomForest"),
    (GradientBoostingRegressor(n_estimators=100, random_state=42), "GradientBoosting"),
    (XGBRegressor(n_estimators=100, random_state=42), "XGBoost"),
    (MLPRegressor(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=42), "NeuralNetwork")
]

# Avaliar cada modelo para Heating Load
print("\nTreinando modelos para Heating Load...")
resultados_heating = []
for modelo, nome in modelos_heating:
    print(f"Treinando {nome}...")
    resultado = avaliar_modelo(modelo, X_train, X_test, y_heating_train, y_heating_test, 
                              nome, "Heating_Load")
    resultados_heating.append(resultado)
    print(f"Concluído. RMSE Test: {resultado['RMSE_Test']:.4f}, R² Test: {resultado['R2_Test']:.4f}")

# Avaliar cada modelo para Cooling Load
print("\nTreinando modelos para Cooling Load...")
resultados_cooling = []
for modelo, nome in modelos_cooling:
    print(f"Treinando {nome}...")
    resultado = avaliar_modelo(modelo, X_train, X_test, y_cooling_train, y_cooling_test, 
                              nome, "Cooling_Load")
    resultados_cooling.append(resultado)
    print(f"Concluído. RMSE Test: {resultado['RMSE_Test']:.4f}, R² Test: {resultado['R2_Test']:.4f}")

# Combinar todos os resultados
todos_resultados = resultados_heating + resultados_cooling
df_resultados = pd.DataFrame(todos_resultados)

# Salvar os resultados
df_resultados.to_csv('../../reports/resultados_modelos.csv', index=False)

print("\nTodos os modelos treinados e avaliados com sucesso!")
display(df_resultados)

# Visualizar os resultados
plt.figure(figsize=(14, 8))

plt.subplot(1, 2, 1)
df_heating = df_resultados[df_resultados['Target'] == 'Heating_Load']
sns.barplot(x='Modelo', y='RMSE_Test', data=df_heating)
plt.title('RMSE - Heating Load')
plt.xticks(rotation=45)
plt.grid(True)

plt.subplot(1, 2, 2)
df_cooling = df_resultados[df_resultados['Target'] == 'Cooling_Load']
sns.barplot(x='Modelo', y='RMSE_Test', data=df_cooling)
plt.title('RMSE - Cooling Load')
plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.savefig('../../reports/figures/comparacao_rmse.png')
plt.show()

# Gráfico de R²
plt.figure(figsize=(14, 8))

plt.subplot(1, 2, 1)
sns.barplot(x='Modelo', y='R2_Test', data=df_heating)
plt.title('R² - Heating Load')
plt.xticks(rotation=45)
plt.grid(True)

plt.subplot(1, 2, 2)
sns.barplot(x='Modelo', y='R2_Test', data=df_cooling)
plt.title('R² - Cooling Load')
plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.savefig('../../reports/figures/comparacao_r2.png')
plt.show()

# Identificar o melhor modelo para cada target
melhor_heating = df_heating.loc[df_heating['R2_Test'].idxmax()]
melhor_cooling = df_cooling.loc[df_cooling['R2_Test'].idxmax()]

print(f"\nMelhor modelo para Heating Load: {melhor_heating['Modelo']}")
print(f"R²: {melhor_heating['R2_Test']:.4f}, RMSE: {melhor_heating['RMSE_Test']:.4f}")

print(f"\nMelhor modelo para Cooling Load: {melhor_cooling['Modelo']}")
print(f"R²: {melhor_cooling['R2_Test']:.4f}, RMSE: {melhor_cooling['RMSE_Test']:.4f}")
```
